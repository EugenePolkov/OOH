```python
def almost_double_factorial(n):
  if n==0:
    return 1
  else:
    l=[x for x in range (0,n+1) if x%2 !=0]
    return reduce (lambda x,y: x*y, l)
```

```python
# Рабочий и полезный декоратор для оценки скорости работы функции
import time
def time_cost_function(fn):
    def wrapper(*args, **kwargs):
        start = time.time()
        fn(*args, **kwargs)
        dt = time.time() - start
        print(f"Время обработки: {dt} сек")
    return wrapper
```

```python
# Автоматическое дифферницрование реализовано через GradientTape
# сначала задаем точки, затем вычисляются значения функции и промежуточные значения простых функций в графе (прямой проход по графу)
# затем идет обратный проход в котором по цепному правилу вычисляются производные во всех промежуточных узлах графа, производные вычисляются только по переменным, по константам - нет
x= tf.Variable (2.0)
y= tf.Variable (5.0)
with tf.GradientTape () as tape:
  f = (x - y)**2 +2*x*y
g = tape.gradient ( f , [x,y])
np.array (g)
```

```python
# бывает что надо вычислять не все частные производные в функции, а только часть, это настраивается параметром watch (сильно экономит память)
x= tf.Variable (2.0)
y= tf.Variable (5.0)
with tf.GradientTape (watch_accessed_variables=False) as tape:
  tape.watch ([x])
  f = (x - y)**2 +2*x*y
g = tape.gradient ( f , [x,y])
np.array (g) # вычислил только производную по x
```

```python
# Gradient Tape поддерживает условные операторы, очень удобно обрабатывать нули или вычислять Хубера
x = tf.Variable(1.0)
with tf.GradientTape() as tape:
    if x < 2.0:
        y = tf.reduce_sum([2.0, 3.0] * x ** 2)
    else:
        y = x ** 2

tape.gradient(y, x)
```

```python
# для успешной работы GradintTape важно несколько правил tape отслеживает только то, что внутри него и то, что tensorflow:
# все промежуточные функции внутри tape и все в явном виде:
x = tf.Variable(1.0)
with tf.GradientTape() as tape:
    y = 2 * x + 1
    z = y ** 2
  # z.assign (y*y) не прокатит, теряется явная зависимость z от y
print (tape.gradient(z, [x,y]))

# не забывать что отслеживаются только var и только tf
for n in range(2):
    with tf.GradientTape() as tape:
        y = x ** 2 + 2 * x # np.square не прокатит
    df = tape.gradient(y, x)
    x.assign (x+1) # х = x+1 не прокатит на втором шаге цикла
    print (df)
```

```python
# простейший градиентный спуск
TOTAL_POINTS = 1000
x = tf.random.uniform(shape=[TOTAL_POINTS], minval=0, maxval=10) # генерируем дату X c шумом
noise = tf.random.normal(shape=[TOTAL_POINTS], stddev=0.2)
k_true = 0.7
b_true = 2.0
y = x * k_true + b_true + noise  # вычисляем у

k = tf.Variable(0.0) # инициализация значений
b = tf.Variable(0.0)
EPOCHS = 500
learning_rate = 0.02

for n in range(EPOCHS):
    with tf.GradientTape() as t:
        f = k * x + b # функция предикта (аналог Wx + b)
        loss = tf.reduce_mean(tf.square(y - f)) # на основе функции предикта -  loss func

    dk, db = t.gradient(loss, [k, b]) # вычисляем градиент

    k.assign_sub(learning_rate * dk) # обновляем веса
    b.assign_sub(learning_rate * db)
print (k.numpy(), b.numpy())  # очень близко к true
```

```python
# встроенный SGD
opt = tf.optimizers.SGD ( learning_rate=0.01)
for n in range(EPOCHS):
    with tf.GradientTape() as t:
        f = k * x + b # функция предикта (аналог Wx + b)
        loss = tf.reduce_mean(tf.square(y - f)) # на основе функции предикта -  loss func

    dk, db = t.gradient(loss, [k, b]) # вычисляем градиент
    opt.apply_gradients (zip ( [dk,db] , [k,b] )) # обновляем веса

print (k.numpy(), b.numpy() )  # очень близко к true
```

Общая формула ЦПТ  - $\overline{x} = \eta\space(\mu, \frac{\sigma^2}{n}) $ то есть выборочное среднее оно принадлежит нормальному распределению, в котором $\mu$ и $\sigma$ связаны с исходным распределением и равны мат ожиданию и дисперсиии исходного распределения при больших N. Эти параметры нам неизвестны, поэтому мы делаем 2 шага:
- принимаем в качестве $\mu$ выборочное  среднее (по ЗБЧ сходится к 𝐄 (x))
- в качестве $\sigma^2$ принимаем либо дисперсию по выборке, либо, если возможно, выводим через 𝚬(x) (binom, poisson, hypergeo, nbinom дисперсия производная от p или λ)   

в нашем случае, мы точно знаем, что это биномиальное распредление (любая доля / процент вида "столько то людей слушают Авторадио", "столько то посетителей совершили покупку" это биномиальное распределение, так как по сути просто серия нулей и единиц, где единица с определенной верятностью возникает)  
поэтому получив $\hat{p}$ =0.4 мы можем вывести $\widehat{Var}=p(1-p)=0.24$  
Соотвественно, выборочное среднее распределено по закону $\overline{x} = \eta\space(0.4, \frac{0.24}{50})$


```python
# дальше мы рассчитываем std и квантили 2,5% слева и справа
std = np.sqrt (0.24/50 )
l_q, r_q = sc.stats.norm.ppf (0.05/2) , sc.stats.norm.ppf (1-0.05/2) # руками
print ('квантили 2,5%', sc.stats.norm.interval (0.95)) #автоматом
#а теперь вычисляем доверительные интервалы 0.95 для распределения с нашими параметрами
print (0.4 + l_q*std, 0.4 + r_q*std) #руками (2 плюс левый граница, а не минус потому, что квантиль уже с минусом идет)
sc.stats.norm.interval (0.95, loc=0.4, scale=std ) #автоматом
# по итогу истинное значение в этот интервал попадает, но сам инетрвал весьма широкий, для сужения нужно брать большую выборку
#ВАЖНО!! Хорошо видно, что ширина ci не зависит от соотношения выборка/генсовокупность, а только от размера выборки, но отношение выборка/генсовокупность важна с другой точки зрения, с точки зрения репрезентативности
```

```python
# Для этого делаем SVD для матрицаы ковариации ХТ*Х, получаем матрицу U
u, d, vh = np.linalg.svd (np.transpose(X) @ X, full_matrices=True )
u, np.transpose(vh)
```

```python
# Отрезаем от матрицы U слева столько колонок сколько нам надо признаков (в данном случае 2)
U_r=u[:,:2]
U_r
```

# Статистика

## Схема матстатистики

**Метод моментов (коротко):**  
1) согласно ЗБЧ принимаем, что $\bar{X}_{n}\approx E(X_{i}) $  
2) Зная из какого распределения пришли $X_{n}$ знаем и формулу $E(X)$ в которой зашиты основные параметры распределения ($p, λ, μ$ и т.п.)  
3) находим параметры подставляя вместо $E(X)$  - $\bar{X}_{n}$ (получаем точечную оценку)



Пусть случайные величины $X_{1}$, ..., $X_{n}$ независимы и одинаково распределены. Закон больших чисел говорит нам, что среднее выборочное $ \bar{X} $ является хорошей оценкой для математического ожидания $E(X_{i})$:

$ \bar{X}_{n} \stackrel{P}{\longrightarrow} E(X_{i}) $

На практике это означает, что при больших $ n $ эти величины равны:

$\bar{X}_{n}\approx E(X_{i}) $

На этой нехитрой идее и построен метод моментов. Как конкретно используется идея, понятно из следующих двух примеров:
 Максим любит киндеры и собирает коллекцию жирОфов. Для этого он покупает шоколадки. Пусть вероятность $p$ --- вероятность того, что в киндере лежит жироф.  
 Максим считает сколько яиц надо купить, чтобы у него появился очередной жироф. Каждый раз Макс записывает номер попытки, с которой у него появилась правильная игрушка. Обозначим эти величины $X_{1}$, ..., $ X_{n} $. Постройте оценку неизвестного параметра $ p $ с помощью метода моментов.

Величины $ X_{i} $ имеют геометрическое распределение, поэтому $ E(X_{i})=\frac{1}{p} $. Принцип метода моментов гласит:  

$\bar{X}_{n}\approx \frac{1}{p}$

Выражаем неизвестный параметр $ p $:
$p\approx \frac{1}{\bar{X}_{n}} $
Это и есть нужная нам оценка:
$\hat{p}_{MM} = \frac{1}{\bar{X}_{n}}$


равномерное  
Допустим, что случайные величины $ X_{1} $, ..., $ X_{n} $ независимы и равномерны на $ [\theta;\theta+1] $. Постройте оценку неизвестного параметра $ \theta $ с помощью метода моментов.


В данном случае $ E(X_{i})=\theta+0.5 $ и, следовательно:    

$ \bar{X}_{n}\approx \theta+0.5 $  
Выражаем $ \theta $:   
$ \theta\approx \bar{X}_{n}-0.5 $  
Это и есть нужная нам оценка:  
$ \hat{\theta}_{MM}=\bar{X}_{n}-0.5 $  


Если говорить более формально...

Определение. Пусть $ X_{i} $ одинаково распределены и независимы, а $ E(X_{i}) $ зависит от неизвестного параметра $ \theta $, скажем $ E(X_{i})=g(\theta) $. Тогда оценкой метода моментов называется случайная величина:

$\hat{\theta}_{MM} = g^{-1}(\bar{X}_{n}) $

Конечно, иногда бывают ситуации, когда математическое ожидание $ E(X_{i}) $ не зависит от $ \theta $. Например, если $ X_{i} $ равномерны на $ [-\theta;\theta] $, то математическое ожидание $ E(X_{i})=0 $. Что делать в такой ситуации?

Неспроста же наш метод называется методом моментов.  Напомним, что $ k $-ым моментом случайной величины $ X_{i} $ называется математическое ожидание $ E(X_{i}^{k}) $

Итак, если условия $\bar{X}_{n}\approx E(X_{i})$ связанного с первым моментом не хватило, то на помощь придет второй момент случайной величины. В силу того же закона больших чисел:

$\frac{\sum X_{i}^{2}}{n} \approx E(X_{i}^{2})$  

равномерное  
Величины $ X_{i} $ независимы и равномерны на $ [-\theta;\theta] $. Постройте оценку неизвестного параметра $ \theta $ с помощью метода моментов.  

Убеждаемся, что $E(X_{i})=0$:  

$E(X_{i}) = \int _{-\theta}^{\theta} x \cdot \frac{1}{2\theta} dx{x} = \left. \frac{x^2}{4 \theta} \right|_{-\theta}^{\theta} = \left( \frac{\theta^2 - (-\theta)^2 }{4 \theta}  \right) =  \frac{\theta^{2} - \theta^{2}}{4 \theta} = 0$

Находим $ E(X_{i}^{2}) $:

$E(X_{i}^{2}) = \int _{-\theta}^{\theta} x^{2} \cdot \frac{1}{2\theta} dx{x} = \left. \frac{x^3}{6 \theta} \right|_{-\theta}^{\theta} = \left( \frac{\theta^3 - (-\theta)^3 }{6\theta}  \right) = \frac{2 \theta^{3}}{6\theta}  = \frac{\theta^{2}}{3}$  

Согласно принципу метода моментов:  

$ \frac{\sum X_{i}^{2}}{n} \approx \frac{\theta^{2}}{3} $  

Выражаем $ \theta $:

$ \theta\approx \sqrt{3\frac{\sum X_{i}^{2}}{n} }$

Это и есть нужная нам оценка:

$ \hat{\theta}_{MM}= \sqrt{3\frac{\sum X_{i}^{2}}{n} } = \sqrt{3 \overline{X^2 } }$

Если не хватит и второго момента, тогда воспользуемся третьим и т.д. Для произвольного $k$ мы имеем:

$\frac{\sum X_{i}^{k}}{n} \approx E(X_{i}^{k})$  

В большинстве случаев хватает именного первого момента. Последующие моменты нужны чаще всего при оценке нескольких параметров.

**Дельта-метод (коротко):**  
Если метод моментов позволяет нам получить какую-то точечную оценку параметра (например, $p ≈ 0.25$), то дельта-метод позволяет получить оценку распределения этого параметра, самое главное - $CI$, то есть по дельта-методу можно определить, что например $p = 0.25 \pm 0.02\ при\ CI = 95\%$   
Уточнение: если речь идет именно о среднем, то его распределение известно через ЦПТ, но если нужно распределение параметра, который не равен $\bar{X}_{n}$, но выражается через него, то используем дельта-метод      
Схема:  
1) согласно методу моментов принимаем, что параметр(ы) распределения как-то выражается через $\bar{X}_{n}$  
2) Записывае интересующий нас параметр в виде функции от $\bar{X}_{n}$ (например $g(x) = pq = \bar{X}_{n}(1-\bar{X}_{n})$  
3) находим производную $g(x)$ (на практике, нужна не общая производная, а ее значение в точке равной оцениваемому параметру, то есть в нашем примере в точке  $\bar{X}_{n}$)  
4) Считаем распределение по формуле: $g(x) \sim \mathcal {N}(g(\mu),\sigma^2 (g'(\mu))^2)$, то есть обычное нормальное распределение, где центр ($loc$) всегда равен точечной оценке параметра ( $\bar{X}_{n}(1-\bar{X}_{n})$ , в нашем примере), а дисперсия это произведение дисперсии нормально распределенного среднего ($\bar{X}_{n}$ в нашем примере) на квадрат значения производной функции $g(x)$ в точке $loc$


Нормальное распределение возникает, если суммируется большое количество независимых одинаково распределенных случайных величин. Однако оно возникает и в других ситуациях! Дельта-метод основан на том факте, что даже нелинейная функция от нормально распределенной случайной величины  иногда имеет распределение близкое к нормальному.

**Откуда берётся дельта-метод**

Если функция $g(t)$ дифференцируема, то в окрестности точки $\mu$ функция $g(t)$ похожа на прямую, то есть

$$
g(t) \approx g(\mu) + g'(\mu) \cdot (t - \mu).
$$

Об этом нам говорит математический анализ, в частности, разложение в ряд Тэйлора.

Линейное преобразование нормально распределенной случайной величины оставляет её нормально распределенной, если угловой коэффициент отличен от нуля, т.е.

$$
g'(\mu) \neq 0.
$$

Если $X \sim \mathcal {N}(\mu, \sigma^2)$ и  дисперсия $X$ мала, то $X$ практически всегда попадает в небольшую окрестность $\mu$, а в ней $f$ похожа на линейную функцию и

$$
g(X) \approx N(\mu, \sigma^2 (g'(\mu)^2).
$$

Получаем практическую версию дельта-метода. Если:

-  $g(t)$ --- дифференциируема;
-  $g'(\mu) \neq 0$;
- $X \sim \mathcal {N}(\mu,\sigma^2)$;
- дисперсия $\sigma^2$ мала;

тогда

$$
g(X) \sim \mathcal {N}(g(\mu),\sigma^2 (g'(\mu))^2).
$$

ЗБЧ позволяет использовать средние в качестве оценок для различных параметров. ЦПТ подсказывает как среднее будет распределено. Однако на практике часто встречаются ситуации, когда оценка параметра --- это функция от среднего.  Дельта-метод позволяет в такой ситуации понять как будет распределена оценка. Полученное распределение можно использовать для строительства доверительного интервала.

**Дельта-метод на практике:**

*Равномерное:*  
Пусть случайные величины $X_1, \ldots, X_{100}$ $iid$ $U[2;8]$. Как будут распределены $\bar{x}$ и $\frac{1}{\bar{x}}$?


С $\bar{x}$ всё будет просто. Воспользуемся ЦПТ, по ней

$$
\bar{x}\sim \mathcal {N} \left( E(X_i), \frac{Var(X_i)}{n} \right).
$$

Для равномерного распределения $E(X_i) = \frac{2 + 8}{2} = 5$, $Var(X_i) = \frac{(8-2)^2}{12} = 3$.

Получается, что среднее посчитанное по сотне наблюдений будет иметь распределение

$$
\bar{x}_{100} \sim \mathcal {N} \left( 5, \frac{3}{100} \right).
$$

Для поиска распределения $\frac{1}{\bar{x}}$ воспользуемся дельта-методом:

$$
g(t) = \frac{1}{t} \qquad g'(t) = -\frac{1}{t^2} \qquad g(\mu) = \frac{1}{5} \qquad g'(\mu) = - \frac{1}{25}.
$$

Остаётся только подставить  найденные значения в формулу и получить, что

$$
\frac{1}{\bar{x}_{100}} \sim \mathcal {N} \left( \frac{1}{5}, \frac{3}{100} \cdot \left(-\frac{1}{25} \right)^2\right).
$$

*Пуассон*  
Пусть $X_1, \ldots, X_n\ iid\ Pois(\lambda)$.   С помощью дельта-метода найдите как распределена оценка вероятности $\mathbb {P} (X_i = 0)$.


В качестве оценки для $\lambda$ будем использовать оценку метода моментов, $\bar{x}$.  Среднее по ЦПТ имеет асимптотически нормальное распределение

$$
\bar{x}\sim \mathcal {N} \left(\lambda, \frac{\lambda}{n} \right).
$$

Вероятность того, что $X_i = k$ считается по формуле

$$
\mathbb {P}(X_i = k) = \frac{\lambda^k}{k!} \cdot e^{-\lambda},
$$

в частности

$$
\mathbb {P}(X_i = 0) = e^{-\lambda}.
$$

Для оценки последней, $e^{-\bar{x}}$ нам нужно найти распределение. Воспользуемся  дельта-методом:

$$
g(t) = e^{-t} \qquad g'(t) = -e^{-t}
$$

Подставим значения в формулу и получим, что

$$
e^{-\bar{x}} \sim \mathcal {N} \left( e^{-\lambda},  \frac{\lambda}{n} \cdot e^{-2 \cdot \lambda}  \right).
$$

В дисперсию можем подставить вместо $\lambda$ её оценку

$$
e^{-\bar{x}} \sim \mathcal {N} \left( e^{-\lambda},  \frac{\bar{x}}{n} \cdot e^{-2 \cdot \bar{x}}  \right).
$$

Такое распределение мы сможем использовать для строительства доверительных интервалов и дальнейшего анализа.


**Дельта-метод в теории**

Естественно, строгая формулировка идеи <<дисперсия $\sigma^2$ мала>> использует понятие предела и последовательностей случайных величин.

Если:  $g(t)$ --- дифференцируема, $g'(\mu)\neq 0$, и последовательность случайных величин $X_1, X_2, \ldots, X_n, \ldots $ удовлетворяет условию:


$\sqrt{n} (X_n - \mu) \overset{d}{\to}  \mathcal {N}(0,\sigma^2)$,


тогда последовательность $g(X_n)$ удовлетворяет условию:

$\sqrt{n} (g(X_n) - g(\mu)) \overset{d}{\to} \mathcal {N}(0,\sigma^2 (g'(\mu))^2 )$

## Оценки (смещение, состоятельность, эффективность)

Когда статистик оценивает какой-нибудь параметр, он хочет:

- несмещённость (при фиксированном размере выборки мы в среднем не ошибаемся)
- состоятельность (при бесконечном расширении выборки мы приходим к истинному значению) *самое важное свойство, если оценка несостоятельна, остальное бессмысленно*
- эффективность (у нашей оценки самая маленькая дисперсия для её класса (например, для всех несмещенных оценок)) *обычно берут несколько оценок с одинаковым смещением и среди них ищут с минимальной дисперсией*
- асимптотическая нормальность оценки

## Алгоритм для получения доверительных интервалов среднего:
- асимптотика, хорошо работает ЦПТ, условия:
  -  много наблюдений (больше 20-30)
  -  в выборке нет выбросов
  -  исходное распределение **ЛЮБОЕ**, но не сильно скошенное

используем нормальное распределение $\bar{x}\sim \mathcal {N} \left(\bar{x}, \frac{Var(X_i)}{n} \right)$, в качестве второго параметра используется выборочная **несмещенная** дисперсия (то есть с $n-1$ в знаменателе) поделенная на размер выборки  
считаем сигму, домножаем на квантили стандартного $\mathcal {N} (0,1)$, получаем левые и правые отступы от  $\bar{x}$     

- точная оценка (распределение Стьюдента):
  -  используется для малого кол-ва наблюдений (на количествах больше 20-30 просто сходится к асимптотике)
  -  исходное распределение только **НОРМАЛЬНОЕ**

оценка получится более широкой чем у асимптотики, логика вся та же самая, что и в асимптотике, только квантили берем не из стандартного $\mathcal {N}$ ,  а из  Стьюдента
$\bar x \pm t_{1 - \frac{\alpha}{2}} \cdot \sqrt{\frac{\hat s^2}{n}}$
также используем выборочную **несмещенную** дисперсию

*Примечание*: это все для случая когда неизвестна исходная дисперсия, если же она известна (к примеру, погрешность прибора) то интервалы строятся от исходной дисперсии (сигмы)
